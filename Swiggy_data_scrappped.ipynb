```python
itemsname=[]
itemsprice=[]
itemsrating=[]
restaurantname=[]
areaname=[]
restaurantrating=[]
timing=[]
DeliveryFee=[]
Distance=[]
Discount=[]

for i in range(len(data1)):
  lists=data1[i]['card']['card']
  if "info" in lists:
    itemsname.append(lists['info']["name"])
    items=lists['info']
    itemsprice.append((int(items.get("price",float())))/100)
    itemsrating.append(items['ratings']['aggregatedRating'].get('rating'))
  if "restaurant" in lists:
    restaurantname.append(lists['restaurant']['info'].get('name'))
    areaname.append(lists['restaurant']['info'].get('areaName'))
    restaurantrating=(lists['restaurant']['info'].get('avgRating'))
    timing.append(lists['restaurant']['info']['sla'].get('slaString'))
    DeliveryFee.append((int(lists['restaurant']['info']['feeDetails'].get('amount')))/100)
    Distance.append(lists['restaurant']['info']['sla'].get("lastMileTravelString"))
    Discount.append(lists['restaurant']['info'].get('aggregatedDiscountInfo'))
    
    scrapeddata={'itemsname':itemsname,'itemsprice':itemsprice,'itemsrating':itemsrating,'restaurantname':restaurantname,'areaname':areaname,'restaurantrating':restaurantrating,'timing':timing,'DeliveryFee':DeliveryFee,'Distance':Distance,'Discount':Discount}
    df=pd.DataFrame(scrapeddata)
    df.to_csv('scraped_data.csv')
```
